Data of UN Volunteers is available in our Github repo. We load the csv file using the genfromtxt method of numpy library 

```
csv = np.genfromtxt ('https://raw.githubusercontent.com/United-Nations-Volunteers/Volunteers-Data/master/data/UNVolunteersData.csv', delimiter=",",usecols=np.arange(0,13),dtype=None, encoding=None)
```
